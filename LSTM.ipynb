```python
from keras.models import Sequential
from keras.layers import Embedding, LSTM, Dense

# Initialize model
model = Sequential()

# Add Embedding layer with explicit input shape
model.add(Embedding(input_dim=1127, output_dim=150, input_shape=(326,)))

# Flatten the output from Embedding layer
model.add(LSTM(500))

# Add Dense output layer for classification
model.add(Dense(1127, activation='softmax'))

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Display model summary
model.summary()
```
